### カーネルが落ちないようにメモリ管理

```python
def print_varsize():
    import types
    print("{}{: >15}{}{: >10}{}".format('|','Variable Name','|','  Size','|'))
    print(" -------------------------- ")
    for k, v in globals().items():
        if hasattr(v, 'size') and not k.startswith('_') and not isinstance(v,types.ModuleType):
            print("{}{: >15}{}{: >10}{}".format('|',k,'|',str(v.size),'|'))
        elif hasattr(v, '__len__') and not k.startswith('_') and not isinstance(v,types.ModuleType):
            print("{}{: >15}{}{: >10}{}".format('|',k,'|',str(len(v)),'|'))
```

### 関数の定義

```python
import numpy as np
import xgboost as xgb
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
from sklearn import datasets, metrics, cross_validation

def objective(params):
    
    params['eta'] = float(params['eta'])
    params['max_depth'] = int(params['max_depth'])
    params['min_child_weight'] = float(params['min_child_weight'])
    params['gamma'] = float(params['gamma'])
    params['subsample'] = float(params['subsample'])
    params['colsample_bytree'] = float(params['colsample_bytree'])
    
    params['objective'] = 'multi:softprob'
    params['eval_metric'] = 'mlogloss'
    params['booster'] = 'gbtree'
    params['num_class'] = 3
    params['silent'] = 1
    params['nthread'] = 2
    params['seed'] = 12345

    skf = cross_validation.StratifiedKFold(
        y_train_opt, # Samples to split in K folds
        n_folds=3, # Number of folds. Must be at least 2.
        shuffle=True, # Whether to shuffle each stratification of the data before splitting into batches.
        random_state=12345 # pseudo-random number generator state used for shuffling
    )

    boost_rounds = []
    score = []

    for train, test in skf:
        _train_x, _test_x, _train_y, _test_y = \
            X_train_opt[train], X_train_opt[test], y_train_opt[train], y_train_opt[test]

        train_xd = xgb.DMatrix(_train_x, label=_train_y)
        test_xd = xgb.DMatrix(_test_x, label=_test_y)
        watchlist = [(train_xd, 'train'),(test_xd, 'eval')]

        model = xgb.train(
            params,
            train_xd,
            num_boost_round=150,
            evals=watchlist,
            early_stopping_rounds=10
        )

        boost_rounds.append(model.best_iteration)
        score.append(model.best_score)

    print('average of best iteration:', np.average(boost_rounds))
    return {'loss': np.average(score), 'status': STATUS_OK}

def optimize(trials):
    space = {
        "objective": "multi:softprob",
        "eval_metric": "mlogloss",
        
        #Control complexity of model
        "eta" : hp.quniform("eta", 0.2, 0.6, 0.05),
        "max_depth" : hp.quniform("max_depth", 1, 10, 1),
        "min_child_weight" : hp.quniform('min_child_weight', 1, 10, 1),
        "gamma" : hp.quniform('gamma', 0, 1, 0.05),
        
        #Improve noise robustness 
        "subsample" : hp.quniform('subsample', 0.5, 1, 0.05),
        "colsample_bytree" : hp.quniform('colsample_bytree', 0.5, 1, 0.05),
        
        'num_class' : 3,
        'silent' : 1
    }
    # minimize the objective over the space
    best_params = fmin(
        fn=objective,
        space=space,
        algo=tpe.suggest,
        trials=trials,
        max_evals=10
    )

    return best_params
```

```python
X_train_opt = X_train.as_matrix()
X_test_opt = X_test.as_matrix()
y_train_opt = y_train.as_matrix()
y_test_opt = y_test.as_matrix()

trials = Trials()
best_params = optimize(trials)
```

```python
from sklearn.svm import SVC
#from hpsklearn import HyperoptEstimator, svc
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import StandardScaler
```

```python
params = {
    'C':hp.loguniform("C", np.log(1), np.log(100)),
    'kernel':hp.choice('kernel',['rbf','poly']),
    'gamma': hp.loguniform("gamma", np.log(0.001), np.log(0.1)),
}
count = 0
```

### 欠損値の補完

```python
med_imp = Imputer(missing_values=0, strategy='median', axis=0)
med_imp.fit(df_X)
df_X = med_imp.transform(df_X)
```

### 標準化

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_Y, random_state=0)

std_scl = StandardScaler()
std_scl.fit(X_train)
X_train = std_scl.transform(X_train)
X_test = std_scl.transform(X_test)
```

### GridSearch

```python
svc_param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1, 10, 100]
}

svc_grid_search = GridSearchCV(SVC(), svc_param_grid, cv=3)
svc_grid_search.fit(X_train, y_train)
```

```python
print('Train score: {:.3f}'.format(svc_grid_search.score(X_train, y_train)))
```

```python
print('Test score: {:.3f}'.format(svc_grid_search.score(X_test, y_test)))
```

```python
print('Confusion matrix:\n{}'.format(confusion_matrix(y_test, svc_grid_search.predict(X_test))))
```

```python
print('Best parameters: {}'.format(svc_grid_search.best_params_))
```

```python
print('Best estimator: {}'.format(svc_grid_search.best_estimator_))
```

```python
def function(args):
    print (args)
    clf = SVC(**args)
    clf.fit(X_train, y_train)
    prediction = clf.predict(X_test)
    global count
    count = count + 1
    score = accuracy_score(y_test, prediction)
    print ("%s回目" % str(count), score)
    return -accuracy_score(y_test, prediction)

best = fmin(function, params, algo=tpe.suggest,max_evals=100)
print ("best estimate parameters", best)

clf = SVC(**best)
print (clf)
```
